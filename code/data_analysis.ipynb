# Data Analysis

Visualization of covid-deaths/cases in eu compared to the pps (adjusted gdp)

# Import Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Prepare death/gdp datapoints

```python
death_data = pd.read_csv('../data/processed_data/combined_death_data.csv')
x = death_data['year_week']
y = death_data['rate_14_day']
s = death_data['pps']

max_pps = s.max()
min_pps = s.min()
min_year = death_data['year'].min()
max_year = death_data['year'].max()
```

# Plot death/gdp as scatterplot

```python
fig, ax = plt.subplots(figsize=(16,9))
scatterplot = ax.scatter(x, y, c=s ,vmin=min_pps, vmax=max_pps)

# add label metadata
ax.set_xticks([i*52 for i in range(max_year-min_year+1)],[str(min_year+i) for i in range(max_year-min_year+1)])
ax.set_xlabel('Years')
ax.set_ylabel('Covid-19 deaths per million (last 14 days)')
fig.colorbar(scatterplot, orientation='vertical', label = "Purchasing power standard (local price adjusted gdp)")

plt.savefig("../data/graphs/covid_death_gdp_scatterplot.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

# Prepare cases/gdp datapoints

```python
cases_data = pd.read_csv('../data/processed_data/combined_cases_data.csv')

x = (cases_data['year'].astype(str) + "-" + cases_data['week'].astype(str))
y = cases_data['rate_14_day']
s = cases_data['pps']

max_pps = s.max()
min_pps = s.min()
min_year = cases_data['year'].min()
max_year = cases_data['year'].max()
```

# Plot cases/gdp as scatterplot

```python
fig, ax = plt.subplots(figsize=(16,9))
scatterplot = ax.scatter(x, y, c=s ,vmin=min_pps, vmax=max_pps)

# add label metadata
ax.set_xticks([i*52 for i in range(max_year-min_year+1)],[str(min_year+i) for i in range(max_year-min_year+1)])
ax.set_xlabel('Years')
ax.set_ylabel('Covid-19 cases per 100.000 (last 14 days)')
fig.colorbar(scatterplot, orientation='vertical', label = "Purchasing power standard (local price adjusted gdp)")

plt.savefig("../data/graphs/covid_cases_gdp_scatterplot.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
